## Data Cleaning

```python
# Convert zip_code and street to string (categorical)

df['zip_code'] = df['zip_code'].astype(str)
df['street'] = df['street'].astype(str)
```

```python
# Drop rows with missing critical data or invalid zero values

df = df.dropna(subset=['price', 'bed', 'bath', 'state', 'brokered_by'])
df = df[(df['price'] > 0) & (df['bed'] > 0) & (df['bath'] > 0)]
```

```python
# Drop irrelevant or too-missing columns

df.drop(columns=['prev_sold_date'], inplace=True)
```

```python
# Convert house_size, acre_lot to numeric

df['house_size'] = pd.to_numeric(df['house_size'], errors='coerce')
df['acre_lot'] = pd.to_numeric(df['acre_lot'], errors='coerce')
```

```python
# Final check

print("New shape after cleaning:", df.shape)
print(df.isnull().sum())
```

```python
# Drop missing city rows

df = df[df['city'].notnull()]
```

```python
# Fill acre_lot missing with median

df['acre_lot'].fillna(df['acre_lot'].median(), inplace=True)
```

```python
# Fill house_size missing with median

df['house_size'].fillna(df['house_size'].median(), inplace=True)
```

```python
print("New shape after cleaning:", df.shape)
print(df.isnull().sum())
```

## Connect to SQLite

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("realtor.db")
cursor = conn.cursor()
```

## Store Raw Data in a Temporary Table

```python
df.to_sql("realtor_raw", conn, if_exists="replace", index=False)
```

## Normalized Tables

```python
cursor.executescript("""
DROP TABLE IF EXISTS Broker;
DROP TABLE IF EXISTS Location;
DROP TABLE IF EXISTS Listing;

CREATE TABLE Broker (
    broker_id REAL PRIMARY KEY
);

CREATE TABLE Location (
    location_id INTEGER PRIMARY KEY AUTOINCREMENT,
    street TEXT,
    city TEXT,
    state TEXT,
    zip_code TEXT
);

CREATE TABLE Listing (
    listing_id INTEGER PRIMARY KEY AUTOINCREMENT,
    broker_id REAL,
    location_id INTEGER,
    status TEXT,
    price REAL,
    bed REAL,
    bath REAL,
    acre_lot REAL,
    house_size REAL,
    FOREIGN KEY (broker_id) REFERENCES Broker(broker_id),
    FOREIGN KEY (location_id) REFERENCES Location(location_id)
);
""")

conn.commit()
```

## Insert Data

```python
# Unique Brokers

cursor.execute("""
INSERT OR IGNORE INTO Broker (broker_id)
SELECT DISTINCT brokered_by
FROM realtor_raw
WHERE brokered_by IS NOT NULL;
""")

conn.commit()
```

```python
# Unique Locations

cursor.execute("""
INSERT INTO Location (street, city, state, zip_code)
SELECT DISTINCT street, city, state, zip_code
FROM realtor_raw;
""")

conn.commit()
```

```python
# Listings Table

cursor.execute("""
INSERT INTO Listing (broker_id, location_id, status, price, bed, bath, acre_lot, house_size)
SELECT 
    r.brokered_by,
    l.location_id,
    r.status,
    r.price,
    r.bed,
    r.bath,
    r.acre_lot,
    r.house_size
FROM realtor_raw r
JOIN Location l
ON  r.street = l.street
AND r.city = l.city
AND r.state = l.state
AND r.zip_code = l.zip_code;
""")

conn.commit()
```

```python
print("Brokers:", pd.read_sql("SELECT COUNT(*) FROM Broker", conn))
print("Locations:", pd.read_sql("SELECT COUNT(*) FROM Location", conn))
print("Listings:", pd.read_sql("SELECT COUNT(*) FROM Listing", conn))
```
